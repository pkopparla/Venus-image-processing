# A step-by-step walkthrough for generating regular gridded data from an Akatsuki UVI image of Venus 
by Pushkar Kopparla


The Akatsuki Level 3 data which provides physical quantites on regular latitude longitude grids is a very convenient dataset for scientific analysis. Here we look into how this gridded data can be extracted from UVI images of Venus and associated geometry data. The main references for the methods used here are this paper: [Ogohara et al., 2017](https://doi.org/10.1186/s40623-017-0749-5) and this technical document by [Ogohara and Murakami](https://darts.isas.jaxa.jp/planet/project/akatsuki/doc/l3/vco_l3_algorithm_v01.pdf). The data repository for UVI data is on [JAXA DARTS](https://darts.isas.jaxa.jp/planet/project/akatsuki/uvi.html.en).

We start from the Akatsuki Level 2 data, which contains calibrated radiances along with geometry information of the observation in FITS format (see the first figure below). For more details about data levels refer to the Ogohara paper above. If you want to execute this notebook the files you will need are:
1. Spice kernels from [here](https://darts.isas.jaxa.jp/doi/vco/vco-00001.html). Within that directory, you need to access the meta-kernel file at this location "vco-v-spice-6-v1.0/vcosp_1000/extras/mk/vco_v05.tm". Make sure to point the "PATH_VALUES" in the meta-kernel file to access the kernels stored at "vco-v-spice-6-v1.0/vcosp_1000/data/". These files contain information about the spacecraft location, time, pointing directions of various instruments and other ancillary information. There are tutorials [here](https://naif.jpl.nasa.gov/naif/tutorials.html) if you want to learn more about these kernels.
2. A Level 2b FITS file: [here](https://data.darts.isas.jaxa.jp/pub/pds3/vco-v-uvi-3-cdr-v1.0/vcouvi_1004/data/l2b/r0144/uvi_20200302_090113_283_l2b_v20.fit)
3. The corresponding Level 3b NetCDF file: [here](https://data.darts.isas.jaxa.jp/pub/pds3/extras/vco_uvi_l3_v1.0/vcouvi_7004/data/l3b/netcdf/r0144/uvi_20200302_090113_283_l3b_v20.nc)

Also you will need several python libraries listed in the last cell of this notebook. Copy paste this list into a `requirements.txt` file and use `conda` or `pip` to install libraries from this file into your python environment. I used python 3.8.10.


First of all, let's have a look at our starting Level 2b image.

```python
from astropy.io import fits
import matplotlib.pyplot as plt

with fits.open("uvi_20200302_090113_283_l2b_v20.fit") as hdul:
    data = hdul[1].data

plt.figure(figsize=(8, 8))
_ = plt.imshow(data, origin="lower", cmap="inferno")
```

To begin handling geometry information, the first thing to do is to load up the Spice kernels, and calculate the sub-spacecraft point on Venus for our observation time. We can check for correctness by matching it against the listed value in the Level 2b data header. We will be using the [Spicepy](https://spiceypy.readthedocs.io/en/main/) module for  Spice routines to extract geometry information. 

```python
import spiceypy as spice
import numpy as np

METAKR = "vco_v05.tm"
spice.furnsh(METAKR)  # load kernels
with fits.open("uvi_20200302_090113_283_l2b_v20.fit") as hdul:
    header = hdul[1].header  # get header from fits file
utctim = header["DATE-OBS"]  # get the date string of the observation
et = spice.str2et(utctim)  # calculate ephemeris time
subpnt, _, obsvec = spice.subpnt(
    "near point: ellipsoid", "Venus", et, "IAU_VENUS", "NONE", "AKATSUKI"
)  # get subpnt from et
reclat = spice.reclat(subpnt)  # convert from rectangular coords to lat lon
print(
    "This is the location of the sub-spacecraft point \
from Spice: %.3f %.3f"
    % (np.degrees(reclat[1]), np.degrees(reclat[2]))
)
print(
    "Listed location of sub-spacecraft point from FITS header:",
    header["S_SSCLON"],
    header["S_SSCLAT"],
)
```

We are able to calculate the latitude and longitude for the sub-spacecraft point correctly. Next, we want to determine the location of the sub-spacecraft point on the camera detector of the UVI instrument. To do this, we use the observation vector calculated in the previous cell and rotate it into the instrument reference frame (it was calculated in the Venus reference frame). Then, assuming that the camera works like a pinhole camera, a relationship can be derived for the pixel location of the point on the camera detector given a viewing direction and the focal length of the camera (see Equation 9 of [Ogohara et al., 2012](http://dx.doi.org/10.1016/j.icarus.2011.05.017)).

There are some subtleties here with how the references frames are stored on file, the following is Dr. Murakami's explanation of how these work. The references for this information are [imagecoord.txt](https://data.darts.isas.jaxa.jp/pub/pds3/vco-v-uvi-3-cdr-v1.0/vcouvi_1004/document/imagecoord.txt), which describes the image coordinate system and [vco_uvi_v12.ti](https://data.darts.isas.jaxa.jp/pub/pds3/vco-v-spice-6-v1.0/vcosp_1000/data/ik/vco_uvi_v12.ti ), which describes relationships between the instrument reference frame and image coordinates. The important figure is Figure 5 in the second document which I copy below for ease of reading:
<pre>
           +Ysc_proj, +Yuvib_proj,
              ^ -Ximg_uvi_raw, +Yimg_fliprot
              |
        ++    |    --
              |           -Zsc_proj,
              x---------> -Xuvib_proj,
         -Xsc, +Zuvib     +Yimg_uvi_raw, +Ximg_fliprot

        ==         **
</pre>
The image coordinates are `X_fliprot` and `Y_fliprot` and `VCO_UVI_BASE` instrument reference frame coordinates are `Xuvib_proj` and `Yuvib_proj`. `-Xuvib_proj` has same direction with `+Ximg_fliprot`, and `+Yubib_proj` has same direction with `+Yimg_fliprot`. Therefore, given that the center of the camera detector is at pixels \[512.5, 512.5\], there are 1024 pixels in each direction and each pixel has a width of 0.013mm we can calculate the pixel line numbers given an \[x,y\] position in millimeters on the detector as follows. 

```python
pform = spice.pxform(
    "IAU_VENUS", "VCO_UVI_283", et
)  # get the rotation matrix from IAU_Venus to instrument frame, this is a 283 nm observation
pos = spice.mxv(
    pform, obsvec
)  # rotate the observation vector for the subspacecraft point into instrument frame
x, y = (
    -pos[0] * 63.3 / pos[2],
    -pos[1] * 63.3 / pos[2],
)  # use pinhole model to find position on CCD, 63.3 mm is the focal length of the camera
spice_x_line = 512.5 + x / 0.013  # convert positions to x and y line numbers
spice_y_line = 512.5 - y / 0.013
print(
    "Actual sub-pnt line numbers from FITS header:",
    header["S_SSCPX"],
    header["S_SSCPY"],
)
print(
    "Sub-pnt line numbers from VCO_UVI_283 frame: %.3f %.3f"
    % (spice_x_line, spice_y_line)
)
```

The above cell uses `VCO_UVI_283` for the reference frame, though the Murakami and Ogohara tech doc mentions `VCO_UVI_BASE`. A recent clarification from Dr. Ogohara explains why it works this way: The diference between these two frames comes from the mounting alignment of the instrument. `VCO_UVI_BASE` is frame with the ideal mounting alignment, and `VCO_UVI_283` or etc. are estimated, actual mounting alignment of the UVI instrument with respect to the ideal spacecraft structure/frame. To determine what is the correct reference frame for a given image, look for `P_ID` field in the header and use that value.

The next step is to generate a regular latitude longitude grid and use the geometric relationships we derived thus far to map radiance values on the UVI detector pixels to planetary coordinates. We use the Spice routine `illumf` to find if the given location on the planet is visible to the spacecraft and return the line of sight vector from the spacecraft to this point. We then use this vector to determine the corresponding pixel line number on the UVI detector and interpolate radiance values from the nearest pixels to estimate the radiance for the given grid point on the planet.

```python
from scipy.interpolate import RegularGridInterpolator

interpdata = RegularGridInterpolator(
    points=(np.arange(1024), np.arange(1024)), values=data, method="linear"
)  # create an interpolator for the radiances on detector pixels

numlat = 144  # setup lat lon grid, I use small number of points for faster compute
numlon = 288  # actual Level 3 data has 1440x2880 points
lats = np.linspace(-90, 90, num=numlat)
lons = np.linspace(0, 360, num=numlon)

zeroarray = np.zeros([numlat, numlon])  # setup empty array to store data
zeroarray[:] = np.NaN

radius = 6051.8 + 70  # cloud tops are 70 km above surface

for latcount, lat in enumerate(lats):  # loop over all grid points
    for loncount, lon in enumerate(lons):
        recs = spice.latrec(
            radius, np.radians(lon), np.radians(lat)
        )  # calculate surface location in rectangular coords
        trgepc, srfvec, _, _, _, visible, _ = spice.illumf(
            "ellipsoid", "Venus", "Sun", et, "IAU_VENUS", "NONE", "AKATSUKI", recs
        )  # get vector from Akatsuki to grid point and visibility flag

        if visible:  # if this point is visible from Akatsuki
            pform2 = spice.pxfrm2(
                "IAU_VENUS", "VCO_UVI_283", trgepc, et
            )  # calculate rotation from Venus frame to instrument frame
            pos2 = spice.mxv(pform2, srfvec)  # rotate vector into instrument frame
            x, y = (
                -pos2[0] * 63.3 / pos2[2],
                -pos2[1] * 63.3 / pos2[2],
            )  # get position on detector in mm
            x_line = 512.5 + x / 0.013  # convert position to pixel line number
            y_line = 512.5 - y / 0.013
            zeroarray[latcount, loncount] = float(
                interpdata([y_line, x_line])
            )  # extract radiance value from interpolator and assign to grid point

plt.figure(figsize=(12, 8))
plt.contourf(lons, lats, zeroarray, cmap="inferno", levels=10)
plt.title("Uncorrected Data Projection Onto a Regular Grid")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
_ = plt.ylim([-90, 90])
_ = plt.xlim([0, 360])
```

Note how the planet's edge does not coincide with the edge of the observed region, clearly seen at the right bottom of the figure. This happens because the pointing of the camera has errors in measurement (see Sec 2 of Ogohara et al., 2017) and the pointing needs to empirically determined by limb fitting from each image for better accuracy. The limb fitting for pointing correction on Akatsuki data is done using a custom algorithm whose performance and error statistics are described in the Ogohara 2012 and 2017 papers. For simplicity, I will use an edge detection function from an existing python computer vision library, [open-cv](https://opencv24-python-tutorials.readthedocs.io/en/latest/py_tutorials/py_setup/py_intro/py_intro.html#intro). Note that this is only for demonstration purposes, for best results please follow the algorithm from the aforementioned papers.

```python
import cv2


def normalize8(I):
    # A small function to convert Float32 data to UInt8
    # Taken from https://stackoverflow.com/a/53236206
    mn = I.min()
    mx = I.max()

    mx -= mn

    I = ((I - mn) / mx) * 255
    return I.astype(np.uint8)


image = normalize8(
    data
)  # normalize the data into integers from 0 to 255 for better edge detection
edges = cv2.Canny(image, 100, 200)  # detect the planetary limb
plt.imshow(edges, cmap="hsv")  # plot the detected limb
plt.title("Detected Planetary Limb")
_ = plt.axis("off")
```

Once the edge is detected, we then fit an ellipse to the detected limb to determine the center of the planet and it's size. We use the open-cv `fitEllipse` function to determine these parameters.

```python
# https://www.programmersought.com/article/27844593027/
y, x = np.nonzero(edges)  # grab the x,y values of the detected edge in separate arrays
edge_list = np.array(
    [[_x, _y] for _x, _y in zip(x, y)]
)  # construct a list of x,y pairs

ellipse = cv2.fitEllipse(edge_list)  # fit an ellipse to the x,y pairs list
fit_x_line, fit_y_line = (
    ellipse[0][0],
    ellipse[0][1],
)  # retrieve the center of the ellipse

# Visualize the ellipse we just fitted
# Draw a ellipse with red line borders of thickness of 2 px onto the image
plt.figure(figsize=(8, 8))

image2 = np.copy(image)  # setup a background image
image2 = cv2.ellipse(
    image2, ellipse, color=(255, 17, 206), thickness=2
)  # draw an ellipse
plt.imshow(
    image2, cmap="inferno", origin="lower"
)  # show the image with ellipse draw on top
plt.plot(
    [spice_x_line],
    [spice_y_line],
    marker="x",
    color="yellow",
    markersize=10,
    label="Spice Center",
    linestyle="None",
)  # show the center of the ellipse derived from Spice routines (this is the same as sub-pnt from earlier)

import xarray as xr

with xr.open_dataset("uvi_20200302_090113_283_l3b_v20.nc") as ds:
    level3_x_line, level3_y_line = (
        ds["D_SSCPX"].values,
        ds["D_SSCPY"].values,
    )  # get the listed limb fitted ellipse center from Level 3 data

plt.plot(
    level3_x_line,
    level3_y_line,
    marker="o",
    color="black",
    markersize=10,
    label="Listed Fit Center",
    linestyle="None",
)  # mark the listed ellipse center by limb fitting using the Ogohara et al algorithm

plt.plot(
    [fit_x_line],
    [fit_y_line],
    marker="x",
    color="lime",
    markersize=10,
    label="This Fit Center",
    linestyle="None",
)  # mark the ellipse center derived from limb fitting using open-cv
print(
    "Center of planetary disk from Spice: {:.2f} {:.2f}".format(
        spice_x_line, spice_y_line
    )
)
print(
    "Center of planetary disk listed in Level 3 data: {:.2f} {:.2f}".format(
        level3_x_line[0], level3_y_line[0]
    )
)
print(
    "Center of planetary disk from this limb fitting: {:.2f} {:.2f}".format(
        fit_x_line, fit_y_line
    )
)

plt.legend()
_ = plt.axis("off")
```

We see that our fit is reasonably close to the listed value, so we can proceed. Now, we must find the rotation matrix that will rotate the observation vector for the sub-spacecraft point from the Spice derived value (variable `pos` above) to the limb fitted value (variable `lvobs` below). This rotation matrix will be used to correct the pointing error. See Sec 2 of the Ogohara and Murakami tech doc for details on this method. 

```python
def rotation_matrix_from_vectors(vec1, vec2):
    """Find the rotation matrix that aligns vec1 to vec2
    :param vec1: A 3d "source" vector
    :param vec2: A 3d "destination" vector
    :return mat: A transform matrix (3x3) which when applied to vec1, aligns it with vec2.
    Taken from: https://stackoverflow.com/a/59204638
    """
    a, b = (vec1 / np.linalg.norm(vec1)).reshape(3), (
        vec2 / np.linalg.norm(vec2)
    ).reshape(3)
    v = np.cross(a, b)
    c = np.dot(a, b)
    s = np.linalg.norm(v)
    kmat = np.array([[0, -v[2], v[1]], [v[2], 0, -v[0]], [-v[1], v[0], 0]])
    rotation_matrix = np.eye(3) + kmat + kmat.dot(kmat) * ((1 - c) / (s ** 2))
    return rotation_matrix


centx, centy, a = (
    ellipse[0][0],
    ellipse[0][1],
    ellipse[1][0],
)  # get the center and semi-major axis of the fitted ellipse, these are line numbers on the detector


centx = (centx - 512.5) * 0.013  # convert line numbers to mm
centy = (512.5 - centy) * 0.013
a = a * 0.013
Lf = 63.3  # this is the camera focal length
l = np.sqrt(centx**2 + centy**2)
alpha = (
    np.arctan(centy / centx) + np.pi
)  # azimuthal angle, see Ogohara and Murakami, Eq. 2
beta = np.pi / 2 - 1.0 / 2.0 * (
    np.arctan((l - a) / Lf) + np.arctan((l + a) / Lf)
)  # elevation angle, Eq. 3
lvobs = [
    np.cos(beta) * np.cos(alpha),
    np.cos(beta) * np.sin(alpha),
    np.sin(beta),
]  # fitted observation vector, Eq. 1
rot_matrix = rotation_matrix_from_vectors(
    pos, lvobs
)  # calculate the corrected rotation matrix
print(np.isclose(spice.mxv(rot_matrix, pos / np.linalg.norm(pos)), lvobs)) #sanity check to see if rotation is correct
```

As before, we map positions on latitude longitude grid to radiances on the detector pixels. This time, there is one additional rotation with the correction matrix for pointing correction.

```python
correctarray = np.zeros([numlat, numlon])
correctarray[:] = np.NaN
radius = 6051.8 + 70
for latcount, lat in enumerate(lats):
    for loncount, lon in enumerate(lons):
        recs = spice.latrec(radius, np.radians(lon), np.radians(lat))
        trgepc, srfvec, _, _, _, visible, _ = spice.illumf(
            "ellipsoid", "Venus", "Sun", et, "IAU_VENUS", "NONE", "AKATSUKI", recs
        )

        if visible:
            pform2 = spice.pxfrm2("IAU_VENUS", "VCO_UVI_283", trgepc, et)
            pos2 = spice.mxv(pform2, srfvec)
            corr = spice.mxv(rot_matrix, pos2) #this is the pointing correction
            x, y = -corr[0] * 63.3 / corr[2], -corr[1] * 63.3 / corr[2]
            x_line = 512.5 + x / 0.013
            y_line = 512.5 - y / 0.013
            correctarray[latcount, loncount] = float(interpdata([y_line, x_line]))

plt.figure(figsize=(12, 8))
plt.contourf(lons, lats, correctarray, cmap="inferno", levels=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Pointing Corrected Data Projection Onto a Regular Grid")

_ = plt.ylim([-90, 90])
_ = plt.xlim([0, 360])
```

Note how the limb lines up with the observed region edge far better than before. This is the expected result. We can do a quick side-by-side comparison with the Level 3 data.

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
ax[0].contourf(lons, lats, correctarray, cmap="inferno", levels=10)
ax[0].set_ylim([-90, 90])
ax[0].set_xlim([0, 360])
ax[0].set_title("This attempt")
with xr.open_dataset("uvi_20200302_090113_283_l3b_v20.nc") as ds:

    ax[1].contourf(
        ds.longitude, ds.latitude, ds.radiance[0, :, :], cmap="inferno", levels=10
    )
    ax[1].set_ylim([-90, 90])
    ax[1].set_xlim([0, 360])
    ax[1].set_title("Actual Level-3 [uvi_20200302_090113_283_l3b_v20]")
```
